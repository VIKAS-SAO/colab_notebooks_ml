```
import numpy as np
from matplotlib import pyplot as plt
import math 

#   making a bar graph 
x = [1,2,3,4,5,6] 
y1 = [2,4,1,5,2,6]
y2 = [5,1,2,8,9,7]
y3 = [1,3,6,2,9,5]
x_axis = np.arange(len(x))
w = .20

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('vikas')
plt.bar(x_axis,y1 , color = 'b'  ,width = w ,  label='first')
plt.bar(x_axis+w,y2  ,width = w ,color='k', label='second')
plt.bar(x_axis-w,y3  , color= 'r' ,width = w ,  label='three' )
plt.legend()
plt.show()
```

**PIE CHARTS**

```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import math 
from collections import Counter
#   making a IE CHART  from a 

plt.style.use('fivethirtyeight')
data =[10,100,80,30]
labels =['1' ,'2','3' , '4']
colors = ['red' ,'blue' , 'green' ,'yellow']
explode =[0,0,.1,0]



plt.pie(data , labels = labels, colors= colors ,explode = explode ,shadow =True , autopct = '%1.1f%%', wedgeprops={'edgecolor':'blue'})

 
plt.tight_layout()
plt.show()


```

**FILLING AREA ON LINE PLOTS**

```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import math 
from collections import Counter
#   making a IE CHART  from a 

plt.style.use('fivethirtyeight')
data = pd.read_csv('/content/drive/My Drive/salaries.csv')
print(data.columns)

age =data['Age']
python = data['Python']
all_devs  = data['All_Devs']
javascript =  data['JavaScript'] 

plt.plot(age , python , label = 'pyhton')
plt.plot(age , all_devs ,linestyle ='--' ,  label = 'all_devs')
plt.plot(age , javascript ,  label = 'javascript')

median_salary = 60000
# plt.fill_between(age , python  , median_salary ,where = (python>median_salary), color = 'blue' , alpha=.15)
# plt.fill_between(age , python  , median_salary ,where = (python<=median_salary), color = 'red' , alpha=.15)
 
plt.fill_between(age , python  , javascript ,where = (python>javascript), color = 'blue' , alpha=.5)
plt.fill_between(age , python  , javascript ,where = (python<=javascript), color = 'red' , alpha=.5)

plt.title('Median salaries')
 
plt.tight_layout()
plt.legend()
plt.show()




```

**HISTOGRAMS**

**SCATTER PLOTS**

```

plt.style.use('seaborn') 
data = pd.read_csv('/content/drive/My Drive/likes_view_ratio.csv')
view_count = data['view_count']
likes= data['likes'] 
ratio= data['ratio']

plt.scatter(view_count , likes , c = ratio  , edgecolor = 'red' ,alpha = .58 )
plt.title('youtube trending videos')
plt.xlabel('view count')
plt.ylabel('likes')
plt.xscale('log')
plt.yscale('log')
cbar=plt.colorbar()
cbar.set_label('ratio likes/disklikes')
  
plt.tight_layout()
plt.show()
```
