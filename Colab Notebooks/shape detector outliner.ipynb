```python
import time
import numpy as np
import pandas as pd
import seaborn as sns
import math
import sys
import copy
import matplotlib.pyplot as plt 
%matplotlib inline
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore') 
from zipfile import ZipFile 
import cv2
import tensorflow as tf
import imutils
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, image=cap.read()
    hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
    lower_yellow = np.array([25,70,120])
    upper_yellow = np.array([30,255,255])
    
    lower_green = np.array([40,70,80])
    upper_green = np.array([70,255,255])  
    
    lower_red= np.array([0,50,120])
    upper_red = np.array([10,255,255]) 
    
    lower_blue = np.array([90,60,0])
    upper_blue = np.array([121,255,255])
    
    lower_skin = np.array([0, 48, 80], dtype = "uint8")
    upper_skin = np.array([20, 255, 255], dtype = "uint8")
    
    mask1 = cv2.inRange(image , lower_yellow,upper_yellow)
    mask2 = cv2.inRange(image , lower_green,upper_green)
    mask3 = cv2.inRange(image , lower_red,upper_red)
    mask4 = cv2.inRange(image , lower_blue,upper_blue)
    mask5 = cv2.inRange(image, lower_skin , upper_skin)
    
    cnt1 = cv2.findContours(mask1 , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE)
    cnt1 =imutils.grab_contours(cnt1)
    
    cnt2 = cv2.findContours(mask2 , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE)
    cnt2 =imutils.grab_contours(cnt2)
    
    
    cnt3 = cv2.findContours(mask3 , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE)
    cnt3 =imutils.grab_contours(cnt3)
    
    cnt4 = cv2.findContours(mask4 , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE)
    cnt4 =imutils.grab_contours(cnt4)
    
    cnt5 = cv2.findContours(mask5 , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE)
    cnt5 =imutils.grab_contours(cnt5)
    
    
    
    #yellow
    for c in cnt1:
        area  = cv2.contourArea(c)
        if area>=300:
            cv2.drawContours(image ,[c] ,-1 ,(255,255,255) ,  2)
            M  =cv2.moments(c)
            cx= int(M['m10']/M['m00'])
            cy= int(M['m01']/M['m00'])
            cv2.circle(image , (cx,cy) , 7,(255,255,255) , -1)
            cv2.putText(image , 'Yellow' , (cx-10,cy-10),cv2.FONT_HERSHEY_COMPLEX  ,1 ,(255,255,255) , 1)
            
   #green
    for c in cnt2:
        area  = cv2.contourArea(c)
        if area>=300:
            cv2.drawContours(image ,[c] ,-1 ,(0,255,0) ,  2)
            M  =cv2.moments(c)
            cx= int(M['m10']/M['m00'])
            cy= int(M['m01']/M['m00'])
            cv2.circle(image , (cx,cy) , 7,(255,255,255) , -1)
            cv2.putText(image , 'Green' , (cx-10,cy-10),cv2.FONT_HERSHEY_COMPLEX  ,1 ,(255,255,255) , 1)
            
   #red
    for c in cnt3:
        area  = cv2.contourArea(c)
        if area>=300:
            cv2.drawContours(image ,[c] ,-1 ,(0,0,255) ,  2)
            M  =cv2.moments(c)
            cx= int(M['m10']/M['m00'])
            cy= int(M['m01']/M['m00'])
            cv2.circle(image , (cx,cy) , 7,(255,255,255) , -1)
            cv2.putText(image , 'Red' , (cx-10,cy-10),cv2.FONT_HERSHEY_COMPLEX  ,1 ,(255,255,255) , 1)
            
   #blue
    for c in cnt4:
        area  = cv2.contourArea(c)
        if area>=300:
            cv2.drawContours(image ,[c] ,-1 ,(255,0,0) ,  2)
            M  =cv2.moments(c)
            cx= int(M['m10']/M['m00'])
            cy= int(M['m01']/M['m00'])
            cv2.circle(image , (cx,cy) , 7,(255,255,255) , -1)
            cv2.putText(image , 'Blue' , (cx-10,cy-10),cv2.FONT_HERSHEY_COMPLEX  ,1 ,(255,255,255) , 1)
            
    #skin
    for c in cnt4:
        area  = cv2.contourArea(c)
        if area>=300:
            cv2.drawContours(image ,[c] ,-1 ,(255,255,255) ,  2)
            M  =cv2.moments(c)
            cx= int(M['m10']/M['m00'])
            cy= int(M['m01']/M['m00'])
#             cv2.circle(image , (cx,cy) , 7,(255,255,255) , -1)
#             cv2.putText(image , 'Blue' , (cx-10,cy-10),cv2.FONT_HERSHEY_COMPLEX  ,1 ,(255,255,255) , 1)

        
    
    

    
    
    
    
    cv2.imshow('image' , image)
    k = cv2.waitKey(5)
    if k==27:
        break
    
    
cap.release()
cv2.destroyAllWindows()
```
