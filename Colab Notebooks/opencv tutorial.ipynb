```python
#edge detection Hough Detection



image   = cv2.imread('sudoku.jpg',-1)
grey =cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
hsv =cv2.cvtColor(image , cv2.COLOR_BGR2HSV)
edges= cv2.Canny(grey ,50,150 , apertureSize=3)
lines=cv2.HoughLines(edges , 1,np.pi/180,200)

for line in lines:
    rho ,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    
    x1= int(x0 +1000*(-b))
    y1= int(y0 +1000*(a))
    x2= int(x0 -1000*(-b))
    y2= int(y0 -1000*(a))
    
    cv2.line(image ,(x1,y1),  (x2,y2) , (0,0,255) , 2)
    
 

    
cv2.imshow('image' ,image)
cv2.imshow('image1' ,edges)

cv2.waitKey(0)
cv2.destroyAllWindows()
 
 
```

```python
#edge detection Hough Detection Probabilistic




image   = cv2.imread('road.jpg',-1)
grey =cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
hsv =cv2.cvtColor(image , cv2.COLOR_BGR2HSV)
edges= cv2.Canny(grey ,100,110 , apertureSize=3)
lines=cv2.HoughLinesP(edges , 1,np.pi/180,100 ,minLineLength = 100, maxLineGap =10)

for line in lines:
    x1,y1,x2,y2 = line[0]
    
    
    cv2.line(image ,(x1,y1),  (x2,y2) , (0,0,255) ,1)
    
 

    
cv2.imshow('image' ,image)
cv2.imshow('image1' ,edges)

cv2.waitKey(0)
cv2.destroyAllWindows()
 
 
```

```python
#edge detection Hough Detection Probabilistic in video

cap  =cv2.VideoCapture('video.mp4')

while True:
    ret , image = cap.read()
    grey =cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
    hsv =cv2.cvtColor(image , cv2.COLOR_BGR2HSV)
    edges= cv2.Canny(grey ,50,150 , apertureSize=3)
    lines=cv2.HoughLinesP(edges , 1,np.pi/180,100 ,minLineLength = 100, maxLineGap =10)
    if lines is None:
        continue
    
        

    for line in lines:
        x1,y1,x2,y2 = line[0]


        cv2.line(image ,(x1,y1),  (x2,y2) , (0,0,255) ,1)

 

    
    cv2.imshow('image' ,image)
    
    
    
    
    
    k = cv2.waitKey(5)
    if k==27:
        break
cap.release()
cv2.destroyAllWindows()



 
```

```python
# harris corner detector

 



image   = cv2.imread('image.jpg',-1)
grey =cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
grey = np.float32(grey)



# hsv =cv2.cvtColor(image , cv2.COLOR_BGR2HSV)
# edges= cv2.Canny(grey ,100,110 , apertureSize=3)

dst =cv2.cornerHarris(grey , 2,3,0.04)
dst = cv2.dilate(dst  , None)
image[dst>0.01*dst.max()] =[0,0,255]


 
    
cv2.imshow('image' ,image)

cv2.waitKey(0)
cv2.destroyAllWindows()
 
 
```

```python
# shi tomasi corner detector




image   = cv2.imread('image.jpg',-1)
grey =cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
corners  = cv2.goodFeaturesToTrack(grey , 100 , 0.01, 10)
corners = np.int0(corners)






# hsv =cv2.cvtColor(image , cv2.COLOR_BGR2HSV)
# edges= cv2.Canny(grey ,100,110 , apertureSize=3)

for i in corners:
    x,y = i.ravel()
    cv2.circle(image , (x,y) , 1,(0,0,255)  , 2)

 
    
cv2.imshow('image' ,image)

cv2.waitKey(0)
cv2.destroyAllWindows()
 
```
